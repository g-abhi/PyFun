```
fname = "hello-world.txt"
```

```
# w for writing to file (will override pre-existing content)
file_object = open(fname, "w")
```

```
file_object.write("Hello World\n")
```

```
file_object.close()
```

```
# a for appending to the end of the file if it exists
with open(fname, "a") as file_object:
    file_object.write("Hello World again")
```

```
# r for reading from file
# with open and closes the file
with open(fname, 'r') as f:
    print(f.read())
```

```
files_dir = os.path.join("/content", "Pictures")
os.makedirs(files_dir, exist_ok=True)
my_nums = range(0, 12)
for i in my_nums:
    fname = f"{i}.jpg"
    file_path = os.path.join(files_dir, fname)
    if os.path.exists(file_path):
        print(f"skipped {fname}")
        continue
    with open(file_path, 'w') as f:
        f.write("Hello World")
```
